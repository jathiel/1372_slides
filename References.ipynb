## Graphs


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Generate 50 random integers from 0 to 9

data = np.random.randint(10, size=50)
data
```

### Frequency Table

```python
data = pd.Series(data)
data.value_counts().sort_index()
```

### Line Graph

```python
sorted_data = data.value_counts().sort_index()
plt.scatter(sorted_data.index, sorted_data)
plt.xlabel('measured quantity')
plt.ylabel('count')
plt.show()
```

### Bar Graph

```python
plt.bar(sorted_data.index, sorted_data)
plt.xlabel('measured quantity')
plt.ylabel('count')
plt.show()
```

### Frequency Polygon

```python
plt.plot(sorted_data.index, sorted_data, marker='o')
plt.xlabel('measured quantity')
plt.ylabel('count')
plt.show()
```

### Pie Chart

```python
values = [10,20,5,67]
my_labels = ['a','b','c','d']
plt.pie(values,labels=my_labels)
plt.show()
```

### Histogram

```python
# Data set of 301 possible different results

data2 = np.random.randint(301,size=100)

# This histogram breaks the data set into 10 different class intervals 
# that are equally spaced

plt.hist(data2,bins=10)
plt.show()
```

## Scipy module for probability computations

[Documentation Link](https://docs.scipy.org/doc/scipy/reference/stats.html#module-scipy.stats)

```python
import scipy as sp
```

### Probability computations for normal random variables.

Let $X$ be a normal random variable with $\mu =\verb|mu|$ and $\sigma =\verb|sd|$

$\verb|sp.stats.norm.cdf|$ computes $P\{X\leq a\}$

```python
sp.stats.norm.cdf(a, loc = mu, scale = sd)
```

$\verb|sp.stats.norm.ppf|$ computes $z_a$ where $P\{X>z_a\}>a$, the $100(1-a)$th percentile

```python
sp.stats.norm.ppf(a, loc = mu, scale = sd)
```

### Probability computations for binomial random variables.

Let $Y$ be a binomial random variable with $n=\verb|n|$ and $p = \verb|p|$.

$\verb|sp.stats.binom.pmf|$ computes $P\{Y= a\}$

```python
sp.stats.binom.pmf(a,n,p)
```

$\verb|sp.stats.binom.cdf|$ computes $P\{Y\leq a\}$

```python
sp.stats.binom.cdf(a,n,p)
```
